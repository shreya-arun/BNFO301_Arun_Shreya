1. Create a function that finds the frequencies of each base found within the BRE regions that were found in the previous function

```python
# Read the contents of the FASTA file
with open(fileName, "r") as file:
    fasta_content = file.read()

# Split the FASTA content into header and sequence
header, sequence = fasta_content.strip().split('\n', 1)
# Remove any leading '>' character from the header
header = header.lstrip('>')

# Now you can set the sequence equal to a variable
sequence_variable = sequence

def printPribnowSummary(sequenceData):
 #use re function
 pattern = re.compile('[CG]{2}[AG]{1}CGCC')
 count = 0
 matches = []
 #1) Count
 for match in re.findall(pattern, sequenceData):
   count += 1
   matches.append(match)
 print("Total matches", len(matches))
 print()
 #2) Base Frequencies
 for i in range(len(matches[0])):
   basecount = {'A':0 ,'T':0 ,'C':0 ,'G':0 }
   for j in matches:
      base = j[i]
      basecount[base] += 1
 #Print for each unique site (the frequency of each BPs)
   for k,v in basecount.items():
     print(k, v/count)
   print()


printPribnowSummary(sequence_variable)
```

2.  Create a function that finds the number of BRE sites that have a TATA site adjacent to it in the genome
 and 2â†’ determine the frequencies of the variable bases within the sites that contain both neighboring sequences

```python
def printBRE_TATASummary(sequenceData):
    # BRE pattern
    pattern_BRE = re.compile('[CG]{2}[AG]{1}CGCC')
    # TATA pattern
    pattern_TATA = re.compile('TA[ATCG]{3}T')

    # Find all matches for BRE and TATA
    matches_BRE = [(match.start(), match.group()) for match in re.finditer(pattern_BRE, sequenceData)]
    matches_TATA = [(match.start(), match.group()) for match in re.finditer(pattern_TATA, sequenceData)]

    # Filter for BRE matches that are near a TATA box
    near_matches = []
    for bre_start, bre_seq in matches_BRE:
        for tata_start, tata_seq in matches_TATA:
            # Check if TATA is within 50 bases after the BRE
            if bre_start < tata_start <= bre_start + 50:
                near_matches.append((bre_seq, tata_seq, bre_start, tata_start))

    print("Total pairs found:", len(near_matches))
    for bre_seq, tata_seq, bre_start, tata_start in near_matches:
        print(f"BRE: {bre_seq} at position {bre_start}, TATA: {tata_seq} at position {tata_start}")

    # Base frequencies
    if near_matches:
        for i in range(len(near_matches[0][0])):  # Using the length of the first BRE match
            basecount = {'A': 0, 'T': 0, 'C': 0, 'G': 0}
            for bre_seq, _, _, _ in near_matches:
                if i < len(bre_seq):  # Check to avoid index errors
                    base = bre_seq[i]
                    basecount[base] += 1
            total_pairs = len(near_matches)
            print(f"Position {i+1} in BRE matches:")
            for base, count in basecount.items():
                print(f"{base}: {count/total_pairs:.2f}")
            print()

printBRE_TATASummary(sequence_variable)
```

3. Create a definition that determines how far apart the BRE and TATA domains are from eachother and the mean distance from the previous code


```python
def printBRE_TATASummary_Distances(sequenceData):
    # BRE pattern
    pattern_BRE = re.compile('[CG]{2}[AG]{1}CGCC')
    # TATA pattern
    pattern_TATA = re.compile('TA[ATCG]{3}T')

    # Find all matches for BRE and TATA
    matches_BRE = [(match.start(), match.group()) for match in re.finditer(pattern_BRE, sequenceData)]
    matches_TATA = [(match.start(), match.group()) for match in re.finditer(pattern_TATA, sequenceData)]

    # Filter for BRE matches that are near a TATA box
    near_matches = []
    distances = []
    for bre_start, bre_seq in matches_BRE:
        for tata_start, tata_seq in matches_TATA:
            # Check if TATA is within 50 bases after the BRE
            if bre_start < tata_start <= bre_start + 50:
                near_matches.append((bre_seq, tata_seq, bre_start, tata_start))
                distances.append(tata_start - bre_start)

    print("Total pairs found:", len(near_matches))
    for bre_seq, tata_seq, bre_start, tata_start in near_matches:
        print(f"BRE: {bre_seq} at position {bre_start}, TATA: {tata_seq} at position {tata_start}")
    print()

    # Calculate and print mean distance if any pairs found
    if distances:
        mean_distance = sum(distances) / len(distances)
        print("Mean distance between BRE and TATA domains:", mean_distance)
        print("Distances between matches:", distances)
    else:
        print("No pairs found within the specified range.")

printBRE_TATASummary_Distances(sequence_variable)
```

3.  Create a function that simulates 100 genomes of the same length of the sample genome

```python
def calculateFraction(genome):
   Acount = 0
   Tcount = 0
   Ccount = 0
   Gcount = 0
   for base in genome:
       if base == 'A':
           Acount += 1
       elif base == 'T':
           Tcount += 1
       elif base == 'C':
           Ccount += 1
       elif base == 'G':
           Gcount += 1


   # total count
   totalCount = Acount + Tcount + Ccount + Gcount
   basecount = { 'A': Acount / totalCount, 'T': Tcount / totalCount,'C': Ccount / totalCount, 'G': Gcount / totalCount }

   return basecount


def simulateGenome(genome):
  BasefreqDict = calculateFraction(genome)
  genomeL = len(genome)
  States = [] #keys from 'BaseFrqDictt'
  values = []  #values
  for k,v in BasefreqDict.items():
    States.append(k)
    values.append(v)
  SimGenome = ''.join(np.random.choice(States, genomeL, values))
  return SimGenome

calculateFraction(sequence_variable)
simulateGenome(sequence_variable)
```
